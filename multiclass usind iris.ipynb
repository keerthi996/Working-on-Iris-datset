```python
import pandas as pd 
import tensorflow as tf 
from tensorflow.keras import layers
```

```python
data=pd.read_csv("Iris.csv",index_col="Id")[:2000]
print(data.head())
data=data.sample(frac=1)
features=data.drop(["Species"],axis=1)
labels=data["Species"]
labels=pd.get_dummies(labels)
features=tf.cast(tf.constant(features),dtype=tf.float32)
labels=tf.cast(tf.constant(labels),dtype=tf.float32)
print(features.shape,labels.shape)
```

```python
model=tf.keras.Sequential([
    layers.Dense(9,activation="relu"),                      
    layers.Dense(6,activation="relu"),
    layers.Dense(3),
    layers.Activation("softmax")
])
```

```python
model.build(input_shape=(2000,4))
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.categorical_crossentropy,
    metrics=["accuracy"]
)
```

```python
model.fit(
    features,
    labels,
    epochs=100
)
```

```python
predictions=(model.predict(features)>0.5).astype(int)
for i in range(50):
    print(features[i],predictions[i],labels[i])
    print(tf.math.reduce_max(labels[i]))
```
